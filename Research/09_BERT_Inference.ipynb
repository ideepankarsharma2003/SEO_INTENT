```python
intent= {
    0: "commercial", 
    1: "informational", 
    2: "local", 
    3: "navigational",
    4: "transactional"
}
import numpy as np
```

```python
for i in features:
    print(f"feature: {i}")
    raw_result= my_reloaded_model(tf.constant(pd.Series(i).values))
    raw_result_softmax= tf.keras.activations.softmax(raw_result)
    max_index= np.argmax(raw_result_softmax[0])
    
    print(f"raw result: {raw_result}")
    print(f"softmax: {raw_result_softmax}")
    
    print(f"maximum index: {max_index}, intent: {intent[max_index]}")
    print()
```

```python
import pandas as pd
dataset= pd.read_csv("data/keyword_intent.csv")
```

```python
true_count=0
false_count=0
import time

for i in dataset.values:
    print(f"feature: {i[0]}")
    raw_result= my_reloaded_model(tf.constant(pd.Series(i[0]).values))
    raw_result_softmax= tf.keras.activations.softmax(raw_result)
    max_index= np.argmax(raw_result[0])
    
    # print(f"raw result: {raw_result}")
    # print(f"softmax: {raw_result_softmax}")
    
    # print(f"maximum index: {max_index}, intent: {intent[max_index]}")
    
    if intent[max_index].lower()==i[1].lower():
        true_count+=1
    else:
        false_count+=1
        print(f"predicted intent: {intent[max_index]}")
        print(f"true intent: {i[1]}")
        # time.sleep(5)
    
    print()
```
